```python
# upload the dataset, convert text file into table  format
import pandas as pd
dataset = pd.read_csv("chat.txt", header=None, on_bad_lines='skip', encoding='utf8')
dataset = dataset.drop(0)
dataset
```

```python
# check what data inside in 2 & 3 columns.
d=dataset[2].value_counts()
d
```

```python
# concat two columns.
dataset[1] = dataset[1].fillna('')+ '' + dataset[2].fillna('')+ '' + dataset[3].fillna('')
dataset[1]
```

```python
dataset = dataset[[0, 1]]
dataset
```

```python
# drop unnesscary rows/records. and name the columns.
dataset = dataset.drop(1)
dataset.columns= ['Date', 'Chat']
dataset
```

```python
# from the chat, split time and chat and also add the column in the dataset.
df1=dataset['Chat'].str.split('-', n=1, expand=True)
dataset['Time'] = df1[0]
```

```python
df1
```

```python
# split 1 column, as name and message. and add new column in the dataset.
df2= df1[1].str.split(':', n=1, expand=True)
dataset['Name']= df2[0]
dataset['Chat']= df2[1]
```

```python
dataset
```

```python
# rearrange the column.
dataset=dataset[['Date','Time', 'Name', 'Chat']]
dataset
```

```python
dataset = dataset[
    dataset['Name'] != "have-rental-homes-2bhk-home-3bhk-iid-1832564386 For the best experience download the OLX App"
]
```

```python
dataset.dropna(inplace=True)
```

```python
# create a model. and get analyze the single value.
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
senti = sid.polarity_scores(dataset['Chat'][5])
senti
```

```python
# add all the value in the dataset table.
dataset['scores'] = dataset['Chat'].apply(lambda commontext:sid.polarity_scores(commontext))
dataset['positive'] = dataset['scores'].apply(lambda score_dict: score_dict['pos'])
dataset['negative'] = dataset['scores'].apply(lambda score_dict: score_dict['neg'])
dataset['compound'] = dataset['scores'].apply(lambda score_dict: score_dict['compound'])
dataset['neutral'] = dataset['scores'].apply(lambda score_dict: score_dict['neu'])
```

```python
# check which chat is positive or negative
dataset['com_score'] = dataset['compound'].apply(lambda c: 'positive' if c >=0 else 'negative')
# create a table for this,
posneg = pd.DataFrame(dataset['com_score'].value_counts())
posneg
```
