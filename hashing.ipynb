```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import HashingVectorizer
```

```python
dataset= pd.read_csv("news.csv", index_col=False)
dataset= dataset.drop('Unnamed: 0', axis=1)
dataset
```

```python
x= dataset['text']
y= dataset['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=53)
```

```python
vectorizer = HashingVectorizer(n_features=2**4, alternate_sign=False)
count_x_train = vectorizer.fit_transform(x_train)
count_x_test = vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(count_x_train, y_train)
y_pred= model.predict(count_x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm= confusion_matrix(y_pred, y_test)
from sklearn import metrics
score = metrics.accuracy_score(y_pred, y_test)
score
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_pred, y_test)
print(report)
```
