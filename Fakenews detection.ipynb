```python
import pandas as pd
import numpy as np
import itertools
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("news.csv", index_col=False)
dataset
```

```python
dataset= dataset.drop("Unnamed: 0", axis=1)
dataset
```

```python
x= dataset['text']
x
```

```python
y= dataset['label']
y
```

```python
# # preprocessing technique (convert words into vector) . here we can use counter vector
# count_vector = CountVectorizer(stop_words='english')
# count_x= count_vector.fit_transform(x)
# count_y= count_vector.transform(y)
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=53)
# print(count_x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=53)
```

```python
# preprocessing technique (convert words into vector) . here we can use counter vector
# input preprocessing technique, so, take only x_train and x_test.
count_vector = CountVectorizer(stop_words='english')
count_x_train= count_vector.fit_transform(x_train)
count_x_test= count_vector.transform(x_test)
```

```python
print(count_x_train)
```

```python
# toget unique value
len(count_vector.get_feature_names_out())
```

```python
# model creation
clf= MultinomialNB()
clf.fit(count_x_train, y_train)
y_pred= clf.predict(count_x_test)
cm= metrics.confusion_matrix(y_pred, y_test)
print(cm)
```

```python
# to find accuracy
score= metrics.accuracy_score(y_pred, y_test)
score
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_pred, y_test)
print(report)
```
