```python
import pandas as pd
dataset= pd.read_csv("whatsapp_output.csv", index_col=False)
dataset
```

```python
# in topic modelling we do preprocessing technique
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm=tfidf.fit_transform(dataset['Chat'])
dtm
```

```python
# create the model
from sklearn.decomposition import NMF
nmf=NMF(n_components=5, random_state=42)
nmf.fit(dtm)
```

```python
# show all topic in for loop
for index,topic in enumerate(nmf.components_):
    results=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
topic_results=nmf.transform(dtm)
dataset["Topic"]=topic_results.argmax(axis=1)
```
